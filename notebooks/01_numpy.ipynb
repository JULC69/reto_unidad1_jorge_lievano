```python
# 1. CARGAR DATASET

import pandas as pd

df = pd.read_csv(
    '../data/ventas_techventas.csv',
    engine='python',
    encoding='latin1'
)
```

```python
# 2. CREAR ARRAYS NumPy
import numpy as np

cantidades = np.array(df['cantidad'])
precios = np.array(df['precio_unitario'])

# print("Cantidades:")
# cantidades
# print("Precios:")
# precios
```

```python
# 3. PRODUCTO MATRICIAL
ingresos_dot = np.dot(cantidades, precios)
print(ingresos_dot)
```

Com el reto precisamente menciona que el dataset contiene:

valores nulos (nan)
productos vacíos
cantidades negativas

y que deben detectarse y tratarse.

Por tanto, antes de calcular el producto escalar debo limpiar los datos.

```python
# 4.0COMPARAR CON MULTIPLICACIÓN ELEMENTO A ELEMENTO

ingresos_vector = cantidades * precios

print(ingresos_vector)
```

La multiplicación elemento a elemento permitió calcular el ingreso bruto individual de cada transacción. Durante el análisis se detectaron valores nulos que generan resultados NaN y una transacción con ingreso negativo causada por una cantidad negativa. Además, se observó una amplia variabilidad en los ingresos, lo que sugiere la coexistencia de productos de diferentes rangos de precio dentro del catálogo. Estos hallazgos justifican la necesidad de una fase posterior de limpieza y transformación de datos antes de realizar análisis de negocio más avanzados.

```python
# 5.1 Estadísticas - Media
np.mean(precios)
```

```python
# 5.2 Estadísticas - Mediana
np.median(precios)
```

```python
# 5.3 Estadísticas - Desviación estándar
np.std(precios)
```

```python
# 5.4 Estadísticas - Percentil
np.percentile(precios,[25,75])
```

Al calcular las estadísticas descriptivas del campo precio_unitario, todas las métricas retornaron NaN. Esto ocurre porque el arreglo contiene valores faltantes (NaN) y las funciones estadísticas estándar de NumPy no los ignoran automáticamente. Este resultado evidencia un problema de calidad de datos que debe corregirse antes de realizar cualquier análisis descriptivo. Como siguiente paso, se eliminarán o tratarán los valores nulos para obtener métricas representativas del catálogo de productos.

## Reflexión sobre CRISP-DM

Durante este reto fue posible identificar claramente las seis fases de la metodología CRISP-DM. La primera fase, Comprensión del Negocio, consistió en entender la necesidad del CEO de TechVentas S.A.S. de analizar las ventas del primer semestre. La segunda fase, Comprensión de los Datos, implicó explorar el dataset e identificar problemas como valores nulos, productos vacíos y cantidades negativas. En la fase de Preparación de los Datos se realizaron tareas de limpieza, transformación y creación de nuevas variables para facilitar el análisis. Posteriormente, en la fase de Modelado, se utilizaron herramientas como NumPy, SQL y Pandas para calcular métricas, realizar consultas y generar resúmenes de información. La fase de Evaluación permitió verificar la calidad de los resultados obtenidos y validar que respondieran a las preguntas de negocio planteadas. Finalmente, la fase de Despliegue se refleja en la generación de archivos, bases de datos SQLite y reportes listos para ser utilizados por la empresa.

Considero que la fase más importante en este reto es la Preparación de los Datos, ya que la calidad de cualquier análisis depende directamente de la calidad de la información utilizada. Si los datos contienen errores, valores faltantes o inconsistencias y estos no son tratados adecuadamente, los resultados obtenidos pueden ser incorrectos y conducir a decisiones equivocadas. Por esta razón, invertir tiempo en comprender y limpiar los datos es fundamental para garantizar análisis confiables y útiles para la organización.
